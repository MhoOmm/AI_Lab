```python
# Vacuum Cleaner Problem

env = {'A': 1, 'B': 1}   # 1 = Dirty, 0 = Clean
pos = 'A'

print("Initial Environment Status:")
print(env)
print("Vacuum Cleaner at:", pos)
print("-" * 40)

while env['A'] == 1 or env['B'] == 1:
    if env[pos] == 1:
        print(f"Room {pos} is Dirty. SUCK action.")
        env[pos] = 0
        print(f"Room {pos} cleaned.")
    else:
        if pos == 'A':
            print("Room A clean. Moving RIGHT.")
            pos = 'B'
        else:
            print("Room B clean. Moving LEFT.")
            pos = 'A'
    print("Environment:", env)
    print("-" * 40)

print("All rooms clean. Goal achieved!")
```
