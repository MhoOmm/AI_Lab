```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if (a > b and a < c) or (a > c and a < b):
    mid = a
elif (b > a and b < c) or (b > c and b < a):
    mid = b
else:
    mid = c

print("Second largest number is:", mid)
```
