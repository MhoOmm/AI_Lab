```python
# Strings properties
# they are immutable
# string are indexed
# they are iterable
```

```python
str = "Hello this is a test"
str2 = 'Hello this is a test'
str3 = """Hello this is a test"""
str4 = '''Hello this is a test'''

print(str,str2,str3,str4)
```

```python
# indexing
str = "Hello this is a test"
print(str[0])
print(str[-1])
print(str[5])
```

```python
str = "akekjdgfjhv"
for i in str:
    print(i)
```

```python
# string functions

str = "kagdkjhvfjhr  vjhagdgajhjhda"
# length
print(len(str))
# slicing
print(str[0:21:2])
print(str[-1:-16:-1])
# concate (can do only with other string)
str2 = "jahghd"
str3 =  str+str2
print(str3)

# membership
i = 'a'
if i in "aeiou":
    print("Vowel")
else:
    print("Consonant")
    
# lowercase uppercase
lo = "lshrghjhgs"
print(lo.upper())

# uppercase
high = "DLAJKFLKHKAHDKHF"
print(high.lower())

# capitalise
cap = "adjhkjfhkj"
print(cap.capitalize())

# Title
tit = "kadjbkh aohlakjd ajsbkjfbv"
print(tit.title())

# swapcase

# find
substr = "aoh"
print(tit.find(substr))

# replace
str = "Python is fun"
print(str.replace("Python", "C++"))

# Split
s= str.split(" ")
print(s)

# Join
result = ",".join(s)
print(result)
```
