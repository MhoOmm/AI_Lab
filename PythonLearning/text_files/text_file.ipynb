```python
# reading all the lines using readline
f = open("sample.txt","r")
while True:
    data = f.readline()
    if data == '':
        break
    else :
        print(data,end = '')
```

```python
# with keyword
with open("sample.txt","a") as f:
    f.write("\nThis is a test")
```

```python
with open("sample.txt","r") as f:
    print(f.read())
```

```python
# loading a big file into the ram and then discarding it using with
# this is helpful as it can take up less computation and if our ram is low

bigl = ["hello\t" for i in range(1000)]
f = open("bigl.txt","w")
f.writelines(bigl)
f.close()
with open("bigl.txt","r") as f:
    chunk = 100
    while len(f.read(chunk)) > 0:
        print(f.read(chunk))
        f.read(chunk)

```

```python
# seek and tell function
with open("sample.txt","r") as f:
    print(f.read(10))
    print(f.tell())
    f.seek(15)
    print(f.read(10))
    print(f.tell())
    
    
```
